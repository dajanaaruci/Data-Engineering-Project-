It was given that the value of streams is NULL when the chart column is "viral50". In order to have room for faster data 
processing we can replace this unrecognizable definition into null values to use in functions or models.

```python
search_string = "viral50"
matching_rows = df[df['chart'].str.contains(search_string, na=False)]
print(matching_rows)
```

```python
count_matching_rows = df['chart'].str.contains("viral50", na=False).sum()

print(f"Total count of rows containing the word 'viral 50': {count_matching_rows}")
```

```python
if df.isna().sum().sum() == 0:
    print("No missing values.")
else:
    print("There are missing values in the data.")
    print("\nMissing values per column:")
    print(df.isna().sum())
```

```python
# Replace values containing 'viral50' with NaN
affected_rows = df['chart'].str.contains("viral50", na=False)
df.loc[affected_rows, 'chart'] = np.nan

# Count the number of affected rows
count_affected_rows = affected_rows.sum()

print(f"Total count of rows where 'viral50' was replaced with NaN: {count_affected_rows}")
print("\nUpdated DataFrame:")
print(df)
```

```python
# Convert date to datetime and extract components
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek
```

*Filling missing values with interpolate method*

Preserves Data Trends: Forward and backward filling can help maintain the trend and continuity of the data, which is especially useful in time series data.

Handles Consecutive NaNs: The limit parameter allows control over how many consecutive NaNs are filled, preventing excessive filling that might distort the data.

limit is Maximum number of consecutive NaNs to fill. Must be greater than 0.
This method fills missing values by propagating the last valid value forward.
This will forward-fill up to 10 consecutive NaN values with the last valid value.


```python
df.ffill(limit=10, inplace=True)
```

After forward-filling, this will back-fill up to 5 consecutive NaN values with the next valid value.
This method fills missing values by using the next valid value to fill backward.

```python
df.bfill(limit=10, inplace=True)
```

Using both ffill and bfill ensures that missing values are filled from both directions—forward and backward. This approach can be useful when you want to ensure data continuity without introducing bias from just one direction.

Removing outliers
By comparing mean with max & mean with std we got that:
  *  In Start_Lat column, std is greater than mean --> outlier.
  *  In Start_Lng column, std is greater than mean --> outlier.
  *  In Severity column, std is greater than mean --> outlier.
  *  In Pressure(in) column, std is greater than mean --> outlier.

```python
columns = ['Start_Lat', 'Start_Lng'] 

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR #Anything below this boundary is considered an outlier.
    upper_bound = Q3 + 1.5 * IQR #Anything above this boundary is considered an outlier.
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)] 
 #filters the DataFrame df to keep only the rows where
 #the value in the current column lies within the lower_bound and upper_bound
```

```python
df.to_csv('bronzelayerUSAccidents.csv', index=False)
```

The target is the feature "severity"
(The level of impact the accident had on the traffic,
1 indicates the least amount of impact.)

```python
# TO MEASURE THE ACCURACY OF DATA AFTER PROCESSING TO PREDICT THE TARGET VARIABLE, THE SEVERITY
# Performing machine learning classification using the RandomForestClassifier on a sampled dataset.
# First, a random sample (10% of the original dataset) is taken and missing values are dropped.
# Categorical columns are then converted into numeric form using LabelEncoder to allow for numerical analysis.
# The features and target are defined, with the Severity column as the target.
# The dataset is split into training and testing sets, using 70% of the data for training and 30% for testing.
# A RandomForestClassifier is initialized and trained on the training data.

sampled_dff = df.sample(frac=0.1, random_state=42).dropna()

categorical_cols = sampled_dff.select_dtypes(include=['object', 'bool']).columns
label_encoder = LabelEncoder()
for col in categorical_cols:
    sampled_dff[col] = label_encoder.fit_transform(sampled_dff[col])
numeric_cols = sampled_dff.select_dtypes(include=['number']).columns
X = sampled_dff[numeric_cols].drop('Severity', axis=1, errors='ignore')
y = sampled_dff['Severity']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
rf_classifier = RandomForestClassifier(n_jobs=-1, random_state=42) 
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

confusion_mat = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(confusion_mat)

class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```

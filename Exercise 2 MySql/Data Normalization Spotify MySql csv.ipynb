```python
def normalize_dates(df1, chunk_size=100000):
    total_rows = len(df1)
    for start in range(0, total_rows, chunk_size):
        end = start + chunk_size
        print(f"Processing rows {start} to {end}")
        chunk = df1.iloc[start:end]
        df1.loc[start:end, 'release_date'] = chunk['release_date'].apply(normalize_date)
        print(f"Processed {end if end < total_rows else total_rows} rows out of {total_rows}")
    return df1

def normalize_date(date):
    if pd.to_datetime(date, errors='coerce', format='%Y-%m-%d') is not pd.NaT:
        return date
    elif pd.to_datetime(date, errors='coerce', format='%m/%d/%Y') is not pd.NaT:
        return pd.to_datetime(date, format='%m/%d/%Y').strftime('%Y-%m-%d')
    elif pd.to_datetime(date, errors='coerce', format='%Y') is not pd.NaT:
        return f"{date}-01-01"
    else:
        return np.nan

df1 = normalize_dates(df1, chunk_size=100000)
```

Remove names with non-ASCII characters

```python
import re

def clean_names(df, column):
    def clean_text(text):
        return re.sub(r'[^\x00-\x7F]+', 'UNKNOWN', text)

    df[column] = df[column].apply(lambda x: clean_text(x) if isinstance(x, str) else x)
    return df

df1 = clean_names(df1, 'name')
```

Remove names with non-ASCII characters

```python
import re

def clean_names(df2, column):
    def clean_text(text):
        return re.sub(r'[^\x00-\x7F]+', 'UNKNOWN', text)

    df2[column] = df2[column].apply(lambda x: clean_text(x) if isinstance(x, str) else x)
    return df2

df2 = clean_names(df2, 'name')
```

```python
#Remove brackets from genres column
def clean_genres(df2, column):
    def clean_genre(genre):
        if pd.isna(genre):
            return genre
        genre_str = str(genre)
        if genre_str.strip('[]').strip() == '':
            return 'Unknown'
        genre_str = re.sub(r'[\[\]\'"]+', '', genre_str)
        return genre_str
    
    df2[column] = df2[column].apply(clean_genre)
    return df2

df2 = clean_genres(df2, 'genres')
```

```python
df2.to_csv('artists_cleaned.csv', index=False)
```
